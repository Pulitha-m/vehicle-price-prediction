```python
import pandas as pd
from mpmath.identification import transforms
```

```python
df = pd.read_csv("../dataset/processed/cleaned_car_price_dataset.csv")
```

```python
df.head(5)
```

Dropping first columns beacuse it contains id and has no effect on target varaibale on predicting

```python
df = df.drop(df.columns[0],axis = 1)
```

```python
df.head(20)
```

Multiplye price columns by 100000 because it was previously on thousands

```python
df['Price'] = df['Price'] * 100000
```

checking unique values in these columns

```python
print(df['POWER WINDOW'].unique())
print(df['POWER MIRROR'].unique())
print(df['POWER STEERING'].unique())
print(df['AIR CONDITION'].unique())

```

Convert these to numerical value of 1 and 0

```python

df['POWER WINDOW'] = df['POWER WINDOW'].map({"Available" : 1,"Not_Available":0})
df['POWER MIRROR'] = df['POWER MIRROR'].map({"Available" : 1,"Not_Available":0})
df['POWER STEERING'] = df['POWER STEERING'].map({"Available" : 1,"Not_Available":0})
df['AIR CONDITION'] = df['AIR CONDITION'].map({"Available" : 1,"Not_Available":0})
```

```python
df.head(5)
```

Since 2 categories in Condition and Leasing columns used binary encoding in here as well

```python
df['Condition'] = df['Condition'].map({"USED":0,"NEW":1})
df['Leasing'] = df['Leasing'].map({"No Leasing":0,"Ongoing Lease":1})
```

```python
df.head(20)
```

Town doesnt usually affect vehicle price and keeping it will increase column count which will overfit some models


```python
df = df.drop("Town", axis=1)
```

```python
df
```

```python
df['Date'].unique()
```

create year listed and date listed columns from the Date posted to check seasonal and yearly price variations

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Year_Listed"] = df["Date"].dt.year
df["Month_Listed"] = df["Date"].dt.month
```

```python
df.head(10)
```

Added a car age column becasue vehicle age affects with vehcile price

```python
df["Car_Age"] = 2025 - df["YOM"]
```

Added a another column to get mileage per year of cars so if car has run more miles in less years

```python
df['Mileage_per_year'] = df['Millage(KM)'] / df['Car_Age']
```

```python
df.head(5)
```

```python
df['Fuel Type'].unique()
```

```python
df['Gear'].unique()
```

encode with sklearn one hot encoder and convert to a datafram and add to original dataframe

```python
from sklearn.preprocessing import OneHotEncoder


ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)  # sparse=False  get numpy array
X_encoded = ohe.fit_transform(df[["Fuel Type"]])


encoded_df = pd.DataFrame(
    X_encoded,
    columns=ohe.get_feature_names_out(["Fuel Type"]),
    index=df.index
)


df = pd.concat([df, encoded_df], axis=1)
```

```python

ohe_brand = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
brand_encoded = ohe_brand.fit_transform(df[["Brand"]])

brand_encoded_df = pd.DataFrame(
    brand_encoded,
    columns=ohe_brand.get_feature_names_out(["Brand"]),
    index=df.index
)

df = pd.concat([df, brand_encoded_df], axis=1)
```

```python
df.head(5)
```

Since 2 categories convert to 1 and 0 using binary encoding

```python
df["Gear"] = df["Gear"].map({"Automatic": 1, "Manual": 0})
```

```python
df.head(20)
```

```python
df['Engine (cc)'].unique()
```

One hot encoding model will create lot of columns more than 1500 1440 unique models  so get freqs of those

```python
df['Brand_Freq'] = df['Brand'].map(df['Brand'].value_counts())
df['Model_Freq'] = df['Model'].map(df['Model'].value_counts())
```

Generate brand and model avg price

```python
df["Brand_AvgPrice"] = df.groupby("Brand")["Price"].transform("mean")
df["Model_AvgPrice"] = df.groupby("Model")["Price"].transform("mean")
```

```python
df.head(20)
```

```python
df["BrandModel_AvgPrice"] = df.groupby(["Brand", "Model"])["Price"].transform("mean")
```

```python
df["BrandModelYear_AvgPrice"] = df.groupby(["Brand", "Model", "YOM"])["Price"].transform("mean")
```

```python
luxury_brands = {"AUDI", "BMW", "MERCEDES", "LEXUS", "JAGUAR", "LAND ROVER", "PORSCHE", "VOLVO"}
df["Luxury_Brand"] = df["Brand"].apply(lambda x: 1 if x.upper() in luxury_brands else 0)
```

```python
df.info()
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

Drop these categorical columns beacsue now they are encoded

```python
df = df.drop(columns=["Brand","Model", "Date", "Fuel Type"])
```
